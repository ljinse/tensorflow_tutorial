```python
import pandas as pd
import tensorflow as tf
```

```python
csv_file = tf.keras.utils.get_file('heart.csv', 'https://storage.googleapis.com/download.tensorflow.org/data/heart.csv')
```

```python
df = pd.read_csv(csv_file)
```

```python
df.head()
```

```python
df.dtypes
```

```python
df['thal'] = pd.Categorical(df['thal'])
```

```python
df['thal'] = df.thal.cat.codes # thal > text to num
```

```python
df
```

### tf.data.Dataset 사용하여 데이터로드

```python
target = df.pop('target') # df에서 target이 빠짐
df
```

```python
dataset = tf.data.Dataset.from_tensor_slices((df.values, target.values)) # (입력, 라벨)

for feat, targ in dataset.take(5):
    print('Features: {}, Target: {}'.format(feat, targ))
```

```python
tf.constant(df['thal'])
```

```python
dataset
```

```python
train_dataset = dataset.shuffle(len(df)).batch(1)
train_dataset
```

```python
def get_compiled_model():
    model = tf.keras.Sequential([
        tf.keras.layers.Dense(10, activation='relu'),
        tf.keras.layers.Dense(10, activation='relu'),
        tf.keras.layers.Dense(1)
    ])

    model.compile(optimizer='adam',
                loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                metrics=['accuracy'])
    return model
```

```python
model = get_compiled_model()
model.fit(train_dataset, epochs=15)
```

```python
inputs = {key: tf.keras.layers.Input(shape=(), name=key) for key in df.keys()}
inputs
```

```python
x = tf.stack(list(inputs.values()), axis=-1) # axis = -1 가장 낮은 차원 뒤쪽
x = tf.keras.layers.Dense(10, activation='relu')(x)
x = tf.keras.layers.Dense(10, activation='relu')(x)
output = tf.keras.layers.Dense(1)(x)

model_func = tf.keras.Model(inputs=inputs, outputs=output)

model_func.compile(optimizer='adam',
                   loss=tf.keras.losses.BinaryCrossentropy(from_logits=True),
                   metrics=['accuracy'])
```

```python
dict_slices = tf.data.Dataset.from_tensor_slices((df.to_dict('list'), target.values)).batch(16)
for dict_slice in dict_slices.take(1): # 열 정보를 남긴다. 어떤 열이었는지
    print(dict_slice)
```

```python
model_func.fit(dict_slices, epochs=15)
```
